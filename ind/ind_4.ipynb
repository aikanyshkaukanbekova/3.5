```python
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots()
plt.figure(figsize=(20, 20))

response = requests.get('https://cdn.botanichka.ru/wp-content/uploads/2022/10/pochemu-menyaetsya-okras-czvetka_1.jpg')
img = Image.open(BytesIO(response.content))

ax.set_title("Цветочек",
            fontsize=18)

ax.imshow(img)
```
